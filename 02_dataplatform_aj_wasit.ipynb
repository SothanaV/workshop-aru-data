```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='year', y='value', data=df_thai) # <-
plt.xticks(rotation=90)
plt.show()
```

```python
X = df_thai[['year']].to_numpy()
X.shape
```

```python
y = df_thai['value'].to_numpy()
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
X = le.fit_transform(X).reshape(-1,1)
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train.shape
```

```python
linear = LinearRegression()
```

```python
linear.fit(X_train, y_train)
```

```python
linear.score(X_test, y_test)
```

```python
x_pred = linear.predict(X)
x_pred
```

```python
df_pred = pd.DataFrame({
    "y":list(x_pred),
    'x':list(X)
})
df_pred
```

```python
df_pred['x'] = df_pred['x'].apply(lambda x: x[0])
df_pred
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='year', y='value', data=df_thai, label="data")
sns.lineplot(x='x', y='y', data=df_pred, color='r', label="predict")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
